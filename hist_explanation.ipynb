```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.api.types import is_numeric_dtype
```

```python
data = pd.read_csv("data/dataset_train.csv")

data.head()
```

```python
slytherin = data[data["Hogwarts House"] == "Slytherin"]
ravenclaw = data[data["Hogwarts House"] == "Ravenclaw"]
gryffindor = data[data["Hogwarts House"] == "Gryffindor"]
hufflepuff = data[data["Hogwarts House"] == "Hufflepuff"]

houses = [slytherin, ravenclaw, gryffindor, hufflepuff]
colors = ["green", "blue", "red", "yellow"]

fig, axes = plt.subplots(ncols=4, nrows=4, figsize=(8, 8))
fig.set_figwidth(40)
fig.set_figheight(40)

features = data.iloc[:, 6:]

j = 0
k = 0
for column in features:
	for i in range(len(houses)):
		axes[j, k].hist(houses[i][column].dropna(), color=colors[i], alpha=0.4)
	
	axes[j, k].legend(["Slytherin", "Ravenclaw", "Gryffindor", "Hufflepuff"])
	axes[j, k].set_title(column)
	if (k == 3):
		k = 0
		j += 1
	else:
		k += 1

fig.tight_layout()

plt.show()
```
